```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/supply_chain_data.csv')

# Display basic information about the dataset
print(df.info())
```

```python
print("\
First few rows of the dataset:")
print(df.head())
```

```python
print("\Summary statistics:")
print(df.describe())

# Check for missing values
print("\
Missing values:")
print(df.isnull().sum())
```

```python
product_type_counts = df['Product type'].value_counts()
plt.figure(figsize=(10, 6))
product_type_counts.plot(kind='bar')
plt.title('Distribution of Product Types')
plt.xlabel('Product Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('product_type_distribution.png')
plt.close()

print("1. Product Type Distribution:")
print(product_type_counts)

# 2. Price vs Revenue Analysis
plt.figure(figsize=(10, 6))
plt.scatter(df['Price'], df['Revenue generated'])
plt.title('Price vs Revenue')
plt.xlabel('Price')
plt.ylabel('Revenue Generated')
plt.tight_layout()
plt.savefig('price_vs_revenue.png')
plt.close()
```

```python
# 3. Top 5 Suppliers by Production Volume
top_suppliers = df.groupby('Supplier name')['Production volumes'].sum().sort_values(ascending=False).head()
print("\
2. Top 5 Suppliers by Production Volume:")
print(top_suppliers)

# 4. Shipping Carrier Analysis
carrier_counts = df['Shipping carriers'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(carrier_counts.values, labels=carrier_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Shipping Carriers')
plt.savefig('shipping_carriers.png')
plt.close()

# 5. Correlation Analysis
correlation_matrix = df.select_dtypes(include=[np.number]).corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.savefig('correlation_heatmap.png')
plt.close()

print("\3. Correlation Analysis: Heatmap saved as 'correlation_heatmap.png'")
```

```python
demo_counts = df['Customer demographics'].value_counts()
print("\
4. Customer Demographics Distribution:")
print(demo_counts)

# 7. Average Lead Time by Location
avg_lead_time = df.groupby('Location')['Lead time'].mean().sort_values(ascending=False)
print("\
5. Average Lead Time by Location:")
print(avg_lead_time)

print("\
Analysis complete. Check the generated PNG files for visualizations.")
```
